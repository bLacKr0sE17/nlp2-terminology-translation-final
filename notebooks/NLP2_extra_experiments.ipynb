```python
import ast
import pandas as pd

path = "/content/nlp2_evidence/results/nlp2_final_qualitative_outputs_500.csv"
df = pd.read_csv(path)

# Use the first 100 examples for a controlled extra experiment
test_df = df.head(100).copy()

def parse_terms(term_string):
    if isinstance(term_string, dict):
        return term_string
    return ast.literal_eval(term_string)

test_df["terms_dict"] = test_df["terms"].apply(parse_terms)

test_df.head()
```

```python
def make_translation_only_prompt(source, terms_dict):
    term_lines = []

    for source_term, target_term in terms_dict.items():
        term_lines.append(f"{source_term} -> {target_term}")

    terms_text = "\n".join(term_lines)

    prompt = f"""Translate the English sentence into German.

Use the provided terminology when the source term appears.
Return only the German translation.
Do not add explanations, introductions, labels, or phrases such as "Here is the translation".

English sentence:
{source}

Terminology:
{terms_text}

German translation:"""

    return prompt
```

```python
def generate_outputs(dataframe, batch_size=4, max_new_tokens=160):
    outputs = []

    for start in range(0, len(dataframe), batch_size):
        batch = dataframe.iloc[start:start + batch_size]

        prompts = []
        for _, row in batch.iterrows():
            prompt = make_translation_only_prompt(row["source"], row["terms_dict"])

            messages = [
                {"role": "user", "content": prompt}
            ]

            chat_text = tokenizer.apply_chat_template(
                messages,
                tokenize=False,
                add_generation_prompt=True
            )

            prompts.append(chat_text)

        inputs = tokenizer(
            prompts,
            return_tensors="pt",
            padding=True,
            truncation=True
        ).to(model.device)

        with torch.no_grad():
            generated = model.generate(
                **inputs,
                max_new_tokens=max_new_tokens,
                do_sample=False,
                temperature=None,
                top_p=None,
                pad_token_id=tokenizer.eos_token_id
            )

        # Decode only newly generated tokens, not the prompt
        for i in range(len(batch)):
            input_length = inputs["input_ids"][i].shape[0]
            new_tokens = generated[i][input_length:]
            text = tokenizer.decode(new_tokens, skip_special_tokens=True).strip()
            outputs.append(text)

        print(f"Generated {len(outputs)} / {len(dataframe)}")

    return outputs

test_df["base_translation_only"] = generate_outputs(test_df, batch_size=4)
test_df[["source", "terms", "base_translation_only"]].head()
```

```python
import sacrebleu

def compute_bleu(predictions, references):
    bleu = sacrebleu.corpus_bleu(
        predictions,
        [references]
    )
    return bleu.score

def compute_term_accuracy(outputs, terms_list):
    total_terms = 0
    matched_terms = 0

    for output, terms_dict in zip(outputs, terms_list):
        output_lower = str(output).lower()

        for source_term, target_term in terms_dict.items():
            total_terms += 1

            if str(target_term).lower() in output_lower:
                matched_terms += 1

    if total_terms == 0:
        return 0

    return matched_terms / total_terms

bleu_translation_only = compute_bleu(
    test_df["base_translation_only"].tolist(),
    test_df["reference"].tolist()
)

term_acc_translation_only = compute_term_accuracy(
    test_df["base_translation_only"].tolist(),
    test_df["terms_dict"].tolist()
)

print("Translation-only prompt, 100 examples")
print("BLEU:", bleu_translation_only)
print("Term accuracy:", term_acc_translation_only)
```

```python
boilerplate_patterns = [
    "Gerne",
    "Hier ist",
    "Übersetzung",
    "deutsche Übersetzung",
    "Die deutsche Übersetzung",
    "Here is",
    "Translation:"
]

def has_boilerplate(text):
    text = str(text)
    for pattern in boilerplate_patterns:
        if pattern in text:
            return True
    return False

num_boilerplate = test_df["base_translation_only"].apply(has_boilerplate).sum()

print("Outputs with boilerplate:", num_boilerplate, "out of", len(test_df))
```

```python
comparison_rows = []

for method in ["base_basic", "base_strong", "lora_basic", "lora_basic_cleaned", "base_translation_only"]:
    bleu = compute_bleu(
        test_df[method].tolist(),
        test_df["reference"].tolist()
    )

    term_acc = compute_term_accuracy(
        test_df[method].tolist(),
        test_df["terms_dict"].tolist()
    )

    boilerplate_count = test_df[method].apply(has_boilerplate).sum()

    comparison_rows.append({
        "method": method,
        "BLEU": bleu,
        "term_accuracy": term_acc,
        "boilerplate_count": boilerplate_count
    })

comparison_df = pd.DataFrame(comparison_rows)
comparison_df
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

second_model_name = "HuggingFaceTB/SmolLM2-1.7B-Instruct"

second_tokenizer = AutoTokenizer.from_pretrained(second_model_name)
second_tokenizer.padding_side = "left"

second_model = AutoModelForCausalLM.from_pretrained(
    second_model_name,
    torch_dtype=torch.float16,
    device_map="auto"
)

print("Loaded:", second_model_name)
print("Device:", second_model.device)
```

```python
def make_second_model_no_term_prompt(source, terms_dict=None):
    prompt = f"""Translate the English sentence into German.
Return only the German translation.

English sentence:
{source}

German translation:"""

    return prompt


def make_second_model_basic_prompt(source, terms_dict):
    term_lines = []

    for source_term, target_term in terms_dict.items():
        term_lines.append(f"{source_term} -> {target_term}")

    terms_text = "\n".join(term_lines)

    prompt = f"""Translate the English sentence into German.
Use the provided terminology when relevant.

English sentence:
{source}

Terminology:
{terms_text}

German translation:"""

    return prompt
```

```python
def make_chat_input(tokenizer, prompt):
    messages = [
        {"role": "user", "content": prompt}
    ]

    try:
        chat_text = tokenizer.apply_chat_template(
            messages,
            tokenize=False,
            add_generation_prompt=True
        )
    except:
        chat_text = prompt

    return chat_text


def generate_with_second_model(dataframe, prompt_function, batch_size=4, max_new_tokens=160):
    outputs = []

    for start in range(0, len(dataframe), batch_size):
        batch = dataframe.iloc[start:start + batch_size]

        prompts = []

        for _, row in batch.iterrows():
            prompt = prompt_function(row["source"], row["terms_dict"])
            chat_text = make_chat_input(second_tokenizer, prompt)
            prompts.append(chat_text)

        inputs = second_tokenizer(
            prompts,
            return_tensors="pt",
            padding=True,
            truncation=True
        ).to(second_model.device)

        with torch.no_grad():
            generated = second_model.generate(
                **inputs,
                max_new_tokens=max_new_tokens,
                do_sample=False,
                pad_token_id=second_tokenizer.eos_token_id
            )

        for i in range(len(batch)):
            input_length = inputs["input_ids"][i].shape[0]
            new_tokens = generated[i][input_length:]
            text = second_tokenizer.decode(new_tokens, skip_special_tokens=True).strip()
            outputs.append(text)

        print(f"Generated {len(outputs)} / {len(dataframe)}")

    return outputs
```

```python
def generate_with_third_model(dataframe, prompt_function, batch_size=4, max_new_tokens=160):
    outputs = []

    for start in range(0, len(dataframe), batch_size):
        batch = dataframe.iloc[start:start + batch_size]

        prompts = []

        for _, row in batch.iterrows():
            prompt = prompt_function(row["source"], row["terms_dict"])

            messages = [
                {"role": "user", "content": prompt}
            ]

            chat_text = third_tokenizer.apply_chat_template(
                messages,
                tokenize=False,
                add_generation_prompt=True
            )

            prompts.append(chat_text)

        inputs = third_tokenizer(
            prompts,
            return_tensors="pt",
            padding=True,
            truncation=True
        ).to(third_model.device)

        with torch.no_grad():
            generated = third_model.generate(
                **inputs,
                max_new_tokens=max_new_tokens,
                do_sample=False,
                pad_token_id=third_tokenizer.eos_token_id
            )

        for i in range(len(batch)):
            input_length = inputs["input_ids"][i].shape[0]
            new_tokens = generated[i][input_length:]
            text = third_tokenizer.decode(new_tokens, skip_special_tokens=True).strip()
            outputs.append(text)

        print(f"Generated {len(outputs)} / {len(dataframe)}")

    return outputs
```

```python
import re
import pandas as pd

def normalize_text(text):
    text = str(text).lower()
    text = text.replace("ä", "a").replace("ö", "o").replace("ü", "u").replace("ß", "ss")
    text = re.sub(r"[^a-zA-Z0-9\s\-]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

def get_relaxed_variants(term):
    term_norm = normalize_text(term)
    variants = set()

    # Original normalized term
    variants.add(term_norm)

    # Remove spaces/hyphens for compound-like matching
    variants.add(term_norm.replace(" ", ""))
    variants.add(term_norm.replace("-", ""))
    variants.add(term_norm.replace(" ", "-"))

    # Simple German/English-ish endings.
    # This is intentionally heuristic.
    endings = ["e", "en", "er", "es", "s", "n"]

    for ending in endings:
        variants.add(term_norm + ending)

    # If term already has common endings, also check a stem-like version
    for ending in endings:
        if term_norm.endswith(ending) and len(term_norm) > len(ending) + 3:
            stem = term_norm[:-len(ending)]
            variants.add(stem)
            for e2 in endings:
                variants.add(stem + e2)

    # Remove very short variants to avoid accidental matches
    variants = {v for v in variants if len(v) >= 4}

    return variants

def relaxed_term_match(output, target_term):
    output_norm = normalize_text(output)
    output_compact = output_norm.replace(" ", "").replace("-", "")

    variants = get_relaxed_variants(target_term)

    for variant in variants:
        variant_compact = variant.replace(" ", "").replace("-", "")

        # Direct phrase match
        if variant in output_norm:
            return True

        # Compound/spacing-insensitive match
        if variant_compact and variant_compact in output_compact:
            return True

    return False

def compute_relaxed_term_accuracy(outputs, terms_list):
    total_terms = 0
    matched_terms = 0

    for output, terms_dict in zip(outputs, terms_list):
        for source_term, target_term in terms_dict.items():
            total_terms += 1

            if relaxed_term_match(output, target_term):
                matched_terms += 1

    if total_terms == 0:
        return 0

    return matched_terms / total_terms
```

```python
def make_basic_prompt(source, terms_dict):
    term_lines = []

    for source_term, target_term in terms_dict.items():
        term_lines.append(f"{source_term} -> {target_term}")

    terms_text = "\n".join(term_lines)

    prompt = f"""Translate the English sentence into German.
Use the provided terminology when relevant.

English sentence:
{source}

Terminology:
{terms_text}

German translation:"""

    return prompt


def make_chat_input_for_model(tokenizer, prompt):
    messages = [
        {"role": "user", "content": prompt}
    ]

    try:
        chat_text = tokenizer.apply_chat_template(
            messages,
            tokenize=False,
            add_generation_prompt=True
        )
    except Exception:
        chat_text = prompt

    return chat_text
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
import pandas as pd
import gc

def run_basic_prompt_model(model_name, output_column, dataframe, batch_size=4, max_new_tokens=160):
    print("Loading model:", model_name)

    tokenizer_local = AutoTokenizer.from_pretrained(model_name)
    tokenizer_local.padding_side = "left"

    if tokenizer_local.pad_token is None:
        tokenizer_local.pad_token = tokenizer_local.eos_token

    model_local = AutoModelForCausalLM.from_pretrained(
        model_name,
        torch_dtype=torch.float16,
        device_map="auto"
    )

    outputs = []

    for start in range(0, len(dataframe), batch_size):
        batch = dataframe.iloc[start:start + batch_size]
        prompts = []

        for _, row in batch.iterrows():
            prompt = make_basic_prompt(row["source"], row["terms_dict"])
            chat_text = make_chat_input_for_model(tokenizer_local, prompt)
            prompts.append(chat_text)

        inputs = tokenizer_local(
            prompts,
            return_tensors="pt",
            padding=True,
            truncation=True
        ).to(model_local.device)

        with torch.no_grad():
            generated = model_local.generate(
                **inputs,
                max_new_tokens=max_new_tokens,
                do_sample=False,
                pad_token_id=tokenizer_local.eos_token_id
            )

        for i in range(len(batch)):
            input_length = inputs["input_ids"][i].shape[0]
            new_tokens = generated[i][input_length:]
            text = tokenizer_local.decode(new_tokens, skip_special_tokens=True).strip()
            outputs.append(text)

        print(f"{model_name}: generated {len(outputs)} / {len(dataframe)}")

    dataframe[output_column] = outputs

    # Clear model from GPU
    del model_local
    del tokenizer_local
    gc.collect()
    torch.cuda.empty_cache()

    print("Finished:", model_name)
    return dataframe
```

```python
def make_qwen3_clean_basic_prompt(source, terms_dict):
    term_lines = []

    for source_term, target_term in terms_dict.items():
        term_lines.append(f"{source_term} -> {target_term}")

    terms_text = "\n".join(term_lines)

    prompt = f"""Translate the English sentence into German.

Use the provided terminology when relevant.
Return only the German translation.
Do not explain your answer.
Do not list the terminology.
Do not include reasoning.

English sentence:
{source}

Terminology:
{terms_text}

German translation:"""

    return prompt


def generate_qwen3_clean(dataframe, batch_size=4, max_new_tokens=160):
    outputs = []

    for start in range(0, len(dataframe), batch_size):
        batch = dataframe.iloc[start:start + batch_size]
        prompts = []

        for _, row in batch.iterrows():
            prompt = make_qwen3_clean_basic_prompt(row["source"], row["terms_dict"])
            messages = [{"role": "user", "content": prompt}]

            try:
                chat_text = qwen3_tokenizer.apply_chat_template(
                    messages,
                    tokenize=False,
                    add_generation_prompt=True,
                    enable_thinking=False
                )
            except TypeError:
                chat_text = qwen3_tokenizer.apply_chat_template(
                    messages,
                    tokenize=False,
                    add_generation_prompt=True
                )

            prompts.append(chat_text)

        inputs = qwen3_tokenizer(
            prompts,
            return_tensors="pt",
            padding=True,
            truncation=True
        ).to(qwen3_model.device)

        with torch.no_grad():
            generated = qwen3_model.generate(
                **inputs,
                max_new_tokens=max_new_tokens,
                do_sample=False,
                pad_token_id=qwen3_tokenizer.eos_token_id
            )

        for i in range(len(batch)):
            input_length = inputs["input_ids"][i].shape[0]
            new_tokens = generated[i][input_length:]
            text = qwen3_tokenizer.decode(new_tokens, skip_special_tokens=True).strip()
            outputs.append(text)

        print(f"Generated {len(outputs)} / {len(dataframe)}")

    return outputs
```

```python
def count_matched_terms(output, terms_dict):
    output_lower = str(output).lower()
    matched = 0
    total = 0

    for source_term, target_term in terms_dict.items():
        total += 1

        if str(target_term).lower() in output_lower:
            matched += 1

    return matched, total


def term_match_ratio(output, terms_dict):
    matched, total = count_matched_terms(output, terms_dict)

    if total == 0:
        return 0

    return matched / total


def select_qwen3b_candidate(row):
    candidates = {
        "qwen3b_no_term": row["qwen3b_no_term"],
        "qwen3b_basic": row["qwen3b_basic"],
        "qwen3b_strong": row["qwen3b_strong"]
    }

    preference = {
        "qwen3b_strong": 3,
        "qwen3b_basic": 2,
        "qwen3b_no_term": 1
    }

    best_name = None
    best_output = None
    best_score = -1

    for name, output in candidates.items():
        score = term_match_ratio(output, row["terms_dict"])

        if score > best_score:
            best_score = score
            best_name = name
            best_output = output

        elif score == best_score:
            if preference[name] > preference[best_name]:
                best_name = name
                best_output = output

    return best_name, best_output, best_score
```

```python
def make_random_terms(terms_dict, all_target_terms):
    random_terms = {}

    for source_term, correct_target in terms_dict.items():
        random_target = random.choice(all_target_terms)

        # Try to avoid accidentally picking the correct target term
        attempts = 0
        while random_target == correct_target and attempts < 20:
            random_target = random.choice(all_target_terms)
            attempts += 1

        random_terms[source_term] = random_target

    return random_terms

full_df["random_terms_dict"] = full_df["terms_dict"].apply(
    lambda terms: make_random_terms(terms, all_target_terms)
)

full_df[["terms_dict", "random_terms_dict"]].head()
```

```python
def make_repair_prompt(source, current_translation, missing_terms):
    term_lines = []

    for source_term, target_term in missing_terms.items():
        term_lines.append(f"{source_term} -> {target_term}")

    missing_text = "\n".join(term_lines)

    prompt = f"""You are revising a German translation.

The current translation is mostly correct, but it is missing required terminology.
Revise the translation so that it includes the required German terms.
Keep the meaning of the original English sentence.
Return only the revised German translation.
Do not add explanations or introductions.

English sentence:
{source}

Current German translation:
{current_translation}

Required missing terminology:
{missing_text}

Revised German translation:"""

    return prompt
```

```python
def generate_repairs_for_missing(dataframe, batch_size=4, max_new_tokens=160):
    repaired_outputs = dataframe["qwen3b_selected"].tolist()

    repair_indices = dataframe.index[dataframe["num_missing_selected"] > 0].tolist()
    print("Repairing", len(repair_indices), "outputs")

    for start in range(0, len(repair_indices), batch_size):
        batch_indices = repair_indices[start:start + batch_size]
        prompts = []

        for idx in batch_indices:
            row = dataframe.loc[idx]
            prompt = make_repair_prompt(
                row["source"],
                row["qwen3b_selected"],
                row["selected_missing_terms"]
            )

            messages = [
                {"role": "user", "content": prompt}
            ]

            chat_text = third_tokenizer.apply_chat_template(
                messages,
                tokenize=False,
                add_generation_prompt=True
            )

            prompts.append(chat_text)

        inputs = third_tokenizer(
            prompts,
            return_tensors="pt",
            padding=True,
            truncation=True
        ).to(third_model.device)

        with torch.no_grad():
            generated = third_model.generate(
                **inputs,
                max_new_tokens=max_new_tokens,
                do_sample=False,
                pad_token_id=third_tokenizer.eos_token_id
            )

        for i, idx in enumerate(batch_indices):
            input_length = inputs["input_ids"][i].shape[0]
            new_tokens = generated[i][input_length:]
            text = third_tokenizer.decode(new_tokens, skip_special_tokens=True).strip()
            repaired_outputs[idx] = text

        print(f"Repaired {min(start + batch_size, len(repair_indices))} / {len(repair_indices)}")

    return repaired_outputs
```

```python
import pandas as pd

def term_success_for_row(output, terms_dict):
    output_lower = str(output).lower()
    results = []

    for source_term, target_term in terms_dict.items():
        success = str(target_term).lower() in output_lower

        results.append({
            "source_term": source_term,
            "target_term": target_term,
            "success": success
        })

    return results


methods_for_consistency = [
    "base_basic",
    "base_strong",
    "qwen3b_basic",
    "qwen3b_strong",
    "qwen3b_selected",
    "qwen3b_selected_repaired"
]

all_rows = []

for method in methods_for_consistency:
    if method not in full_df.columns:
        print("Skipping missing method:", method)
        continue

    for idx, row in full_df.iterrows():
        term_results = term_success_for_row(row[method], row["terms_dict"])

        for result in term_results:
            all_rows.append({
                "method": method,
                "example_id": idx,
                "source_term": result["source_term"],
                "target_term": result["target_term"],
                "success": result["success"]
            })

term_level_df = pd.DataFrame(all_rows)
term_level_df.head()
```

```python
per_term_df = (
    term_level_df
    .groupby(["method", "source_term", "target_term"])
    .agg(
        occurrences=("success", "count"),
        successes=("success", "sum"),
        success_rate=("success", "mean")
    )
    .reset_index()
)

# Only focus on terms that occur more than once
repeated_terms_df = per_term_df[per_term_df["occurrences"] > 1].copy()

repeated_terms_df.head()
```

```python
consistency_rows = []

for method in repeated_terms_df["method"].unique():
    method_df = repeated_terms_df[repeated_terms_df["method"] == method]

    avg_repeated_success = method_df["success_rate"].mean()
    perfect_consistency = (method_df["success_rate"] == 1.0).mean()
    num_repeated_terms = len(method_df)

    consistency_rows.append({
        "method": method,
        "num_repeated_terms": num_repeated_terms,
        "avg_repeated_term_success": avg_repeated_success,
        "perfect_consistency_rate": perfect_consistency
    })

consistency_summary_df = pd.DataFrame(consistency_rows)
consistency_summary_df
```

```python
def make_qwen3b_definition_prompt(source, terms_dict):
    term_lines = []

    for source_term, target_term in terms_dict.items():
        term_lines.append(
            f'- English term: "{source_term}" | Required German term: "{target_term}". Use this German term exactly when translating the English term.'
        )

    terms_text = "\n".join(term_lines)

    prompt = f"""Translate the English sentence into German.

You are given terminology constraints. Use the required German terms exactly when the corresponding English terms appear.
Return only the German translation. Do not add explanations or introductions.

English sentence:
{source}

Terminology constraints:
{terms_text}

German translation:"""

    return prompt
```

```python
def generate_qwen3b_definition(dataframe, batch_size=4, max_new_tokens=160):
    outputs = []

    for start in range(0, len(dataframe), batch_size):
        batch = dataframe.iloc[start:start + batch_size]
        prompts = []

        for _, row in batch.iterrows():
            prompt = make_qwen3b_definition_prompt(row["source"], row["terms_dict"])

            messages = [
                {"role": "user", "content": prompt}
            ]

            chat_text = third_tokenizer.apply_chat_template(
                messages,
                tokenize=False,
                add_generation_prompt=True
            )

            prompts.append(chat_text)

        inputs = third_tokenizer(
            prompts,
            return_tensors="pt",
            padding=True,
            truncation=True
        ).to(third_model.device)

        with torch.no_grad():
            generated = third_model.generate(
                **inputs,
                max_new_tokens=max_new_tokens,
                do_sample=False,
                pad_token_id=third_tokenizer.eos_token_id
            )

        for i in range(len(batch)):
            input_length = inputs["input_ids"][i].shape[0]
            new_tokens = generated[i][input_length:]
            text = third_tokenizer.decode(new_tokens, skip_special_tokens=True).strip()
            outputs.append(text)

        print(f"Generated {len(outputs)} / {len(dataframe)}")

    return outputs
```
