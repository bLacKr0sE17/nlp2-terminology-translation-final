```python
import os
import ast

os.makedirs("my_consistency_inputs", exist_ok=True)

def parse_terms(x):
    if isinstance(x, dict):
        return x
    if pd.isna(x):
        return {}
    if isinstance(x, str):
        try:
            value = ast.literal_eval(x)
            if isinstance(value, dict):
                return value
        except Exception:
            return {}
    return {}

if "terms_dict" not in df.columns:
    df["terms_dict"] = df["terms"].apply(parse_terms)

systems = [
    "base_basic",
    "base_strong",
    "qwen3b_basic",
    "qwen3b_strong",
    "qwen3b_selected",
    "qwen3b_selected_repaired"
]

available_systems = []

for col in systems:
    if col in df.columns:
        available_systems.append(col)
        df[col].fillna("").astype(str).to_csv(
            f"my_consistency_inputs/{col}.txt",
            index=False,
            header=False
        )

df["source"].fillna("").astype(str).to_csv(
    "my_consistency_inputs/source.txt",
    index=False,
    header=False
)

df["reference"].fillna("").astype(str).to_csv(
    "my_consistency_inputs/reference.txt",
    index=False,
    header=False
)

df[["source", "reference", "terms_dict"]].to_json(
    "my_consistency_inputs/terms.jsonl",
    orient="records",
    lines=True,
    force_ascii=False
)

print("Available systems:", available_systems)
```

```python
import os
import json
import ast
import pandas as pd

os.makedirs("data/submissions/track1", exist_ok=True)

def parse_terms(x):
    if isinstance(x, dict):
        return x
    if pd.isna(x):
        return {}
    if isinstance(x, str):
        try:
            value = ast.literal_eval(x)
            if isinstance(value, dict):
                return value
        except Exception:
            return {}
    return {}

if "terms_dict" not in df.columns:
    df["terms_dict"] = df["terms"].apply(parse_terms)

def write_official_style_jsonl(system_col, filename):
    path = f"data/submissions/track1/{filename}"

    with open(path, "w", encoding="utf-8") as f:
        for _, row in df.iterrows():
            record = {
                "en": str(row["source"]),
                "de": str(row[system_col]),
                "terms": row["terms_dict"]
            }
            f.write(json.dumps(record, ensure_ascii=False) + "\n")

    print("Wrote:", path)

systems_to_write = [
    ("qwen3b_strong", "qwen3b_strong.ende.proper.jsonl"),
    ("qwen3b_selected", "qwen3b_selected.ende.proper.jsonl"),
    ("qwen3b_selected_repaired", "qwen3b_selected_repaired.ende.proper.jsonl"),
]

for col, filename in systems_to_write:
    if col in df.columns:
        write_official_style_jsonl(col, filename)
    else:
        print("Missing column:", col)
```

```python
import ast
import json
import os
import pandas as pd

def force_parse_terms(x):
    if isinstance(x, dict):
        return x

    if pd.isna(x):
        return {}

    if isinstance(x, str):
        # Try Python dict string first: "{'a': 'b'}"
        try:
            value = ast.literal_eval(x)
            if isinstance(value, dict):
                return value
        except Exception:
            pass

        # Try JSON dict string: '{"a": "b"}'
        try:
            value = json.loads(x)
            if isinstance(value, dict):
                return value
        except Exception:
            pass

    return {}

# Always force-parse, even if terms_dict already exists
if "terms_dict" in df.columns:
    df["terms_dict_fixed"] = df["terms_dict"].apply(force_parse_terms)
else:
    df["terms_dict_fixed"] = df["terms"].apply(force_parse_terms)

print(type(df.loc[0, "terms_dict_fixed"]))
print(df.loc[0, "terms_dict_fixed"])
```

```python
os.makedirs("data/submissions/track1", exist_ok=True)

def write_official_style_jsonl(system_col, filename):
    path = f"data/submissions/track1/{filename}"

    with open(path, "w", encoding="utf-8") as f:
        for _, row in df.iterrows():
            terms_obj = row["terms_dict_fixed"]

            record = {
                "src": str(row["source"]),
                "source": str(row["source"]),
                "src_raw": str(row["source"]),
                "mt": str(row[system_col]),
                "target": str(row[system_col]),
                "mt_raw": str(row[system_col]),
                "reference": str(row["reference"]),
                "terms": terms_obj
            }

            f.write(json.dumps(record, ensure_ascii=False) + "\n")

    print("Wrote:", path)

systems_to_write = [
    ("qwen3b_strong", "qwen3b_strong.ende.proper.jsonl"),
    ("qwen3b_selected", "qwen3b_selected.ende.proper.jsonl"),
    ("qwen3b_selected_repaired", "qwen3b_selected_repaired.ende.proper.jsonl"),
]

for col, filename in systems_to_write:
    if col in df.columns:
        write_official_style_jsonl(col, filename)
    else:
        print("Missing column:", col)
```
